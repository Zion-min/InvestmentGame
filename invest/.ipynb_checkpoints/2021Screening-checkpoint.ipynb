```python
# setting

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pykrx import stock
```

```python
stocks = pd.read_excel('/workspace/AlgorithmTrading/2020-2021종가.xlsx', index_col=0)
```

```python
per = pd.read_excel('/workspace/AlgorithmTrading/2020-2021PER.xlsx',index_col=0)
```

```python
pbr = pd.read_excel('/workspace/AlgorithmTrading/2020-2021PBR.xlsx',index_col=0)
```

```python
volt = pd.read_excel('/workspace/AlgorithmTrading/2020-2021베타.xlsx',index_col=0)
```

```python
cap = pd.read_excel('/workspace/AlgorithmTrading/2020-2021시가총액.xlsx',index_col=0)
```

```python
stocks
```

```python
#1개월(영업일20일) 모멘텀 내림차순

revenue = stocks.pct_change(20)
mom = revenue.iloc[-1].sort_values()
```

```python
#변동성이 낮은순

volts = volt.sort_values(by='beta')
volts
```

```python
# PER 낮음
# 소형주
# 저PBR
pers = per.iloc[-1].sort_values()
caps = cap.iloc[-1]
pbrs = pbr.iloc[-1].sort_values()
```

```python
# dataframe 생성, PER은 중요한 스코어이기에 없는 것들은 dropna로 처리
screen = pd.DataFrame([mom, pers, caps, pbrs]).transpose()
screen.columns = ['mom','per','cap','pbr']
screen = screen.dropna()
screen['volt']=volts 
#volt를 나중에 넣어주는 것은 이미 얘는 잘 정렬 되어 있는 상태이기 때문에 transpose 하기전에 넣어주면 안됨!
screen
```

```python
#변동성이 낮은 것으로 정렬 & 평균보다 높은 것들은 제외
screen['volt_rank']=screen['volt'].rank()
screen = screen.sort_values(by='volt_rank')
screen = screen[screen['volt']<=screen['volt'].mean()]
screen
```

```python
# 6개월 모멘텀 내림차순 & 평균보다 높은 것들은 제외
screen['mom_rank']=screen['mom'].rank()
screen = screen.sort_values(by='mom')
screen = screen[screen['mom']<=screen['mom'].mean()]
screen
```

```python
# 저PBR, 저PER & 평균보다 높은 것들은 제외
screen['PBR_rank']=screen['pbr'].rank()
screen['PER_rank']=screen['per'].rank()
screen = screen[screen['pbr']<=screen['pbr'].mean()]
screen = screen[screen['per']<=screen['per'].mean()]
screen
```

```python
# 여러 컬럼을 기준으로 정렬 (중요도 순)
# 시가 총액이 1000억원 이상/ 인 중소형주(30조 이하)
screen = screen.sort_values(by=['volt_rank','mom_rank','PBR_rank','PER_rank'])
screen = screen[(screen['cap']<=30140000000000) & (screen['cap']>=100000000000)]
screen
```

```python
# rank의 평균을 매기고, 평균 내림차순 정렬 
rankmean = screen.iloc[:,5:].mean(axis=1)
screen['mean']=rankmean
screen = screen.sort_values(by=['mean','cap'])[:50]
screen
```
